## Minkowski Distance

The simulation for Minkowski Distance is performed for 'n' number of attributes or dimensions. 'r' is a parameter.

The general formula for Minkowski Distance between two points $ \mathbf{p} = (p_1, p_2, ..., p_n) $ and $ \mathbf{q} = (q_1, q_2, ..., q_n) $ in an n-dimensional space is given by:

$$ D(\mathbf{p}, \mathbf{q}) = \left( \sum_{i=1}^{n} |p_i - q_i|^r \right)^{\frac{1}{r}} $$

Where \(r\) is the order parameter. When \(r = 1\), it reduces to Manhattan distance, and when \( r = 2 \), it becomes Euclidean distance.

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
def unit_circle(parameter):
    theta = np.linspace(0, 2*np.pi, 1000)          # theta ranges from 0 degree to 360 degrees with (360/1000) degree difference
    coordinates = np.zeros((1000,2))               # creating a list of 1000 2-D array for coordinates (x, y)
    for i, angle in enumerate(theta):
        #Converting into polar coordinates form (x= acos(u) and y= asin(u), where a = 1 = unit distance)
        x = np.cos(angle)           
        y = np.sin(angle)
        norm = (np.abs(x)**r + np.abs(y)**r)**(1/r)     # calculating the Minkowski distance for 'r' parameter
        coordinates[i, 0] = x / norm
        coordinates[i, 1] = y / norm
    return coordinates

# Choosing values of r and creating a list
r_values = [0.25, 0.354, 0.5, 0.707, 1, 2, 3, 4, 500]

# Creating the plot
fig, ax = plt.subplots(3, 3, figsize=(15, 15))
ax = ax.flatten()
for ax, r in zip(ax, r_values):
    coordinates = unit_circle(r)
    ax.plot(coordinates[1:999, 0], coordinates[1:999, 1], label=f'r = {r:.2f}') 
    """Not using of first and last coordinates to avoid the elongated tail"""
    ax.legend()
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title(f'Minkowski Unit Circle for r = {r:.2f}')
    ax.axis('equal')
    ax.grid(True)
    ax.set_xlim([-1, 1])
    ax.set_ylim([-1, 1])

plt.tight_layout()
plt.show()

```

When the value of 'r' is taken less than 1, then an elongated tail occurs at the right side. As for r<1, the Minkowski distance doesn't obey the Triangle Inequality, so the distance metric emphasizes certain directions over others (the direction with starting and ending coordinate of the plot.) This leads to a non-uniform scaling of the axes.
